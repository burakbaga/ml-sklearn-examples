 ## PCA 
PCA genellikle çok özelliğimiz olan datalarda kullanılır. Bunların hepsini modele feedlemek yerine bunların kombinasyonları kullanarak daha anlamlı bir veri seti oluştururak. Overfit olma durumununda önüne geçeriz <br>
Elimizde ki veriyi görselleştirmek için de pca kullanarak çok boyutu 2 veya 3 boyuta indirgeyerek görüntüleme yapabiliriz<br>
PCA aynı zamanda eğitimin hızını artırmak içinde kullanılır. 100x100 bir image i 100 feature düşürebiliriz.,

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
data = load_iris()
X = data.data
y= data.target
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
```

```python
pDf = pd.DataFrame(data=principalComponents,columns=["pc1","pc2"])
pDf.head()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)

targets = ["Setosa","Versicolor","Virginica"]
ax.scatter(principalComponents[:,0],principalComponents[:,1],c=y)
plt.show()
```
